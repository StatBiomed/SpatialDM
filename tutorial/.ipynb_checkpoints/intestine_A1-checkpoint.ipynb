## Spatial Clustering of Local Spots

```python
import NaiveDE
import SpatialDE
```

```python
# SpatialDE code
bin_spots = A1.selected_spots.astype(int)[A1.n_spots>=10]

results = SpatialDE.run(A1.spatialcoord, bin_spots.transpose())
```

```python
histology_results, patterns = SpatialDE.aeh.spatial_patterns(A1.spatialcoord, bin_spots.transpose(),
                                                             results, C=6, l=300,
                                                             verbosity=1)
```

```python
plt.figure(figsize=(18/4*3,8))
for i in range(6):
    plt.subplot(2, 3, i + 1)
    plt.scatter(A1.spatialcoord['x'], A1.spatialcoord['y'], marker = 's',c=patterns[i], s=35);
    plt.axis('equal')
    plt_util('Pattern {} - {} genes'.format(i, histology_results.query('pattern == @i').shape[0] ))
#     plt.colorbar(ticks=[])
#     plt.savefig('SpatialDE_patterns.pdf') #2 # new dfb
```

```python
dic=dict()
for i in histology_results.sort_values('pattern').pattern.unique():
    dic['Pattern_{}'.format(i)]=histology_results.query('pattern == @i').sort_values('membership')['g'].values
```

```python
compute_pathway(A1, dic=dic)
```

## Validate EGF upstream downstream molecules

```python
exp = A1.exp.copy()
exp = exp.loc[:, ['RALA','MAPK3','RALB']]
exp.loc[:,['Moran_selected']] = (A1.local_z_p.loc['AREG_EGFR_ERBB2']<0.1).values  # AREG as an example
```

```python
EGF_df = pd.DataFrame(np.vstack((exp.loc[:,['RALA','Moran_selected']].values,
 exp.loc[:,['MAPK3','Moran_selected']].values,
    exp.loc[:,['RALB','Moran_selected']].values)), columns=['expression', 'Moran_selected'])
n_spots=A1.spatialcoord.shape[0]
EGF_df.loc[:,'gene'] = ['RALA']* n_spots+['MAPK3']*n_spots+['RALB']*n_spots
EGF_df.gene=EGF_df.gene.astype('category')
EGF_df.expression=EGF_df.expression.astype(np.float64)
EGF_df.loc[:,'raw_count']=np.expm1(EGF_df.expression)
```

```python
import seaborn as sns
ax = sns.violinplot(x="gene", y="raw_count", hue="Moran_selected",
                     data=EGF_df, palette="muted", split=False)
plt.savefig('egf_downstream.pdf')
```

```python
plt.figure(figsize=(10,10))
for i,m in enumerate(A1.exp.columns[A1.exp.columns.str.contains('MAPK')]):
    plt.subplot(5,5,i+1)
    plt.scatter(A1.spatialcoord.x, A1.spatialcoord.y, c=A1.exp.loc[:,m])
    plt_util(m)
```

## permutation-zscore consistency

```python
x = -np.log10(A1.global_res.z_pval.values)
y = -np.log10(A1.global_res.perm_pval)

x = np.where(x>2, 2, x)
y = np.where(np.isinf(y), 2, y)

corr_plot(x, y, method='spearman')
```
